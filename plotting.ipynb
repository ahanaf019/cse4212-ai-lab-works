```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

plt.figure(figsize=(10,7))
mpl.rcParams['font.size'] = 13
mpl.rcParams['font.family'] = 'serif'
# Data from the results table
models = ['VGG16$^1$', 'VGG16$^2$', 'VGG16$^3$']
accuracy = [0.6375, 0.8062, 0.7375]
precision = [0.7143, 0.9180, 0.7778]
recall = [0.4878, 0.6829, 0.6829]

# Bar width
bar_width = 0.2

# Set up the positions for bars on X-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plots
plt.bar(r1, accuracy, width=bar_width, label='Accuracy')
plt.bar(r2, precision, width=bar_width, label='Precision')
plt.bar(r3, recall, width=bar_width, label='Recall')

# Add labels and title
plt.xlabel('Models', fontdict={'size': 16})
plt.xticks([r + bar_width for r in range(len(models))], models)
plt.ylabel('Scores', fontdict={'size': 16})
plt.ylim([0, 1.05])
# plt.title('Performance Comparison of VGG16 Models')
font_dict = {'size': 11}
plt.legend(font_dict=font_dict)

for i in range(len(models)):
    plt.text(r1[i], accuracy[i] + 0.01, f'{accuracy[i]:.4f}', ha='center', fontdict=font_dict)
    plt.text(r2[i], precision[i] + 0.01, f'{precision[i]:.4f}', ha='center', fontdict=font_dict)
    plt.text(r3[i], recall[i] + 0.01, f'{recall[i]:.4f}', ha='center', fontdict=font_dict)

# Show the plot
plt.show()
```
