# Import Dependencies

```python
import keras
from keras.layers import Dense, MaxPooling2D, Conv2D, Flatten, AveragePooling2D
from keras import Input
from keras.applications import vgg16
from keras.datasets.mnist import load_data
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

# VGG16 Model From Keras

```python
vgg16_model = vgg16.VGG16()
vgg16_model.summary()
```

# CNN which exactly looks like VGG16

```python
inputs = Input(shape=(224,224,3))

x = Conv2D(filters=64, kernel_size=(3,3), 
           padding='same', activation='relu')(inputs)
x = Conv2D(filters=64, kernel_size=(3,3),
           padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=128, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=128, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=256, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3),
           padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Flatten()(x)
x = Dense(4096, activation='relu')(x)
x = Dense(4096, activation='relu')(x)
x = Dense(1000, activation='softmax')(x)

custom_vgg16 = keras.Model(inputs, x)
custom_vgg16.summary()
```

## Setting weights of Custom VGG16 model to ImageNet weights

```python
for i in range(len(custom_vgg16.layers)):
    # print(custom_vgg16.layers[i].name)
    custom_vgg16.layers[i].set_weights(vgg16_model.layers[i].get_weights())
```

# Build a CNN which will look like VGG16, but will not be exactly VGG16

```python
inputs = Input(shape=(256,256,3))

x = Conv2D(filters=256, kernel_size=(5,5), padding='same', activation='relu')(inputs)
x = Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=(1,1), padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=256, kernel_size=(5, 5), padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=(3, 3), padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=(1, 1), padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=512, kernel_size=(5, 5), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 1), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(1, 3), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3), padding='same', activation='relu')(x)
x = MaxPooling2D()(x)


x = Conv2D(filters=512, kernel_size=(5, 5), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 1), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(1, 3), padding='same', activation='relu')(x)
x = Conv2D(filters=512, kernel_size=(3, 3), padding='same', activation='relu')(x)
x = AveragePooling2D()(x)


x = Conv2D(filters=1024, kernel_size=(5, 5), padding='same', activation='relu')(x)
x = Conv2D(filters=1024, kernel_size=(3, 1), padding='same', activation='relu')(x)
x = Conv2D(filters=1024, kernel_size=(1, 3), padding='same', activation='relu')(x)
x = Conv2D(filters=1024, kernel_size=(3, 3), padding='same', activation='relu')(x)
x = AveragePooling2D()(x)


x = Flatten()(x)
x = Dense(2048, activation='relu')(x)
x = Dense(1536, activation='relu')(x)
x = Dense(1536, activation='relu')(x)
outputs = Dense(1000, activation='softmax')(x)

vgg_like = keras.Model(inputs, outputs)
vgg_like.summary()
```
