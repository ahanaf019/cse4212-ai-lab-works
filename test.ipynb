```python
import keras
import tensorflow as tf
```

```python
input = keras.Input(shape=(4))
x = keras.layers.Dense(3)(input)
x = keras.layers.Dense(3)(x)
output = keras.layers.Dense(1)(x)

model = keras.Model(inputs=input, outputs=output)
model.summary()
```

```python
input = keras.Input(shape=(32,32,3))
x = keras.layers.Conv2D(6, (3,3), padding='same')(input)
# x = keras.layers.BatchNormalization()(x)
x = keras.layers.MaxPool2D()(x)
output = keras.layers.Conv2D(2, (3,3), padding='same')(x)
# output = keras.layers.Dense(1)(x)

model = keras.Model(inputs=input, outputs=output)
model.summary()
```

```python
input = keras.Input(shape=(128,64,3))

x = keras.layers.Conv2D(16, (3,3), strides=(2,2), padding='same')(input)
x = keras.layers.MaxPool2D()(x)

x = keras.layers.Conv2D(32, (3,3), padding='same')(x)
x = keras.layers.MaxPool2D()(x)

x = keras.layers.Conv2D(64, (3,3), padding='same')(x)
x = keras.layers.MaxPool2D()(x)

x = keras.layers.Conv2D(128, (3,3), padding='same')(x)
x = keras.layers.MaxPool2D()(x)

x = keras.layers.Flatten()(x)
output = keras.layers.Dense(2)(x)


model = keras.Model(inputs=input, outputs=output)
model.summary()
```

```python
keras.utils.plot_model(model, show_layer_names=True, show_shapes=True)
```

```python
input = keras.Input(shape=(1))
x = keras.layers.Dense(32, activation="relu")(input)
x = keras.layers.Dense(64, activation="relu")(x)
x = keras.layers.Dense(128, activation="relu")(x)
output = keras.layers.Dense(1)(x)

model = keras.Model(inputs=input, outputs=output)
model.summary()
```
