```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 50)
```

```python
w = 2.1
c = -1.4
y1 = w * x + c

plt.figure(figsize=(8,8))
plt.plot(x, y1, '.-')
plt.grid('on')
plt.title(f'y = {w} * x + {c}')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
c = 0.45
y1 = c * np.exp(x)

plt.figure(figsize=(8,8))
plt.plot(x, y1, '.-')
plt.grid('on')
plt.title(f'y = {c} * e^x')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
a = 5
b = -8
c = -7
d = 1
y1 = a * x**3 + b * x**2 + c * x + d

plt.figure(figsize=(8,8))
plt.plot(x, y1, '.-')
plt.grid('on')
plt.title(f'y = {a} * x^3 + {b} * x^2 + {c} * x + {d}')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
